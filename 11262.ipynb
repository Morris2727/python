```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
xL = np.array([1,3,5,7])
yL = np.array([4,6,8,10])
X,Y = np.meshgrid(xL,yL)
```

```python
Z = np.random.randint(1,3,(4,4))
```

```python
plt.contour(X,Y,Z)
```

```python
Z = np.random.randint(1,3,(4,4))
plt.contour(X,Y,Z)
plt.scatter(X.ravel(),Y.ravel(),c=Z.ravel())
```

```python
Z = np.random.randint(1,3,(4,4))
plt.contourf(X,Y,Z)
plt.scatter(X.ravel(),Y.ravel(),c=Z.ravel())
```
